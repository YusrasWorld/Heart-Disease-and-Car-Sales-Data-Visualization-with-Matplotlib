```python
# Create a Python dictionary of 3 of your favourite foods with The keys of the dictionary should be the food name and the values their price
Favorite_food = {'Pizza':3,
                 'Cake': 5,
                 'Icecream':7}
```

```python
# Create a bar graph where the x-axis is the keys of the dictionaryand the y-axis is the values of the dictionary
fig,ax = plt.subplots()
ax.barFavorite_food.keys()),Favorite_food.values())

# Add a title, xlabel and ylabel to the plot
ax.set(title = 'Favorite_food',xlabel = 'Food',ylabel = 'Price')
plt.show()
```

```python
# Make the same plot as above, except this time make the bars go horizontal (Way 1)
fig,ax = plt.subplots()
ax.barh(list(Favorite_food.keys()),list(Favorite_food.values()))

ax.set(title = 'Favorite_food',xlabel = 'Food',ylabel = 'Price')
plt.show()
```

```python
# Make the same plot as above, except this time make the bars go horizontal (way 2 )
fig,ax = plt.subplots()
ax.barh(Favorite_food.keys(),Favorite_food.values())

ax.set(title = 'Favorite_food',xlabel = 'Food',ylabel = 'Price')
plt.show()
```

```python
# Import the '../data/car-sales.csv' into a DataFame called car_sales and view
Car_Sales = pd.read_csv('rolls-royce.csv')
Car_Sales
```

```python
#Check Car_Sales dtypes
Car_Sales.dtypes
```

```python
# Clean Price (USD)
# Convert to string first, then remove all non-digit characters using raw string
Car_Sales['Price (USD)'] = Car_Sales['Price (USD)'].astype(str).str.replace(r'[^\d]', '', regex=True)

# Convert cleaned string to numeric
Car_Sales['Price (USD)'] = Car_Sales['Price (USD)'].astype(float).fillna(0).astype(int)

# Clean Odometer (km)
Car_Sales['Odometer (km)'] = Car_Sales['Odometer (km)'].astype(str).str.replace(r'[^\d]', '', regex=True)
Car_Sales['Odometer (km)'] = Car_Sales['Odometer (km)'].astype(float).fillna(0).astype(int)
```

```python
# Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column
Car_Sales['Total Sales'] = Car_Sales['Price (USD)'].cumsum()
# Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today)
Car_Sales['Sale Date'] = pd.date_range("13/1/2020", periods=len(Car_Sales))
# View the car_sales DataFrame
Car_Sales
```

```python
# Use the plot() function to plot the 'Sale Date' column versus the 'Total Sales' column
plt.plot(Car_Sales['Sale Date'],Car_Sales['Total Sales'])
plt.show()
```
